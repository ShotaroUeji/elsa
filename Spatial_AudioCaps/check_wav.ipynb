```python
import soundfile as sf
import matplotlib.pyplot as plt

stereo, fs = sf.read("cho/test/107897/stereo.wav")  # shape = (T, 2)
plt.figure(figsize=(12, 4))
plt.plot(stereo[:10000, 0], label="Left")
plt.plot(stereo[:10000, 1], label="Right", alpha=0.7)
plt.legend()
plt.title("stereo.wav (L/R waveform sample)")
plt.show()
```
